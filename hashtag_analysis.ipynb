```python
import pandas as pd
import re
import matplotlib.pyplot as plt

# Assuming 'train' is a pandas DataFrame containing 'label' and 'tweet' columns
train = pd.read_csv('train.csv')

hashtags_0 = {}
hashtags_1 = {}

for index, row in train.iterrows():
    label = row['label']
    tweet = row['tweet']

    hashtags = re.findall(r'#(\w+)', tweet)

    if label == 0:
        for hashtag in hashtags:
            hashtags_0[hashtag] = hashtags_0.get(hashtag, 0) + 1
    elif label == 1:
        for hashtag in hashtags:
            hashtags_1[hashtag] = hashtags_1.get(hashtag, 0) + 1

# Get the top 10 hashtags for label 0
top_10_0 = sorted(hashtags_0.items(), key=lambda x: x[1], reverse=True)[:10]
hashtags_0, counts_0 = zip(*top_10_0)

# Get the top 10 hashtags for label 1
top_10_1 = sorted(hashtags_1.items(), key=lambda x: x[1], reverse=True)[:10]
hashtags_1, counts_1 = zip(*top_10_1)

# Create bar plots
plt.figure(figsize=(10, 6))

plt.subplot(121)
plt.bar(hashtags_0, counts_0)
plt.title('Top 10 Hashtags ')
plt.xlabel('Hashtags')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

#plt.subplot(122)
plt.bar(hashtags_1, counts_1,color="pink")
#plt.title('Top 10 Hashtags for Label 1')
plt.xlabel('Hashtags')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
